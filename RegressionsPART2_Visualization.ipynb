#### Function for Later Use

```python
# Function that gives you a list of the dataframe's current tickers
def ticker_list(df):
    x = 1
    tickers=[df.TICKER[0]]
    while x < len(df):
        if df.TICKER[x]!= df.TICKER[x-1]:
            tickers.append(df.TICKER[x])
            x+=1
        else:
            x+=1
    return tickers
```

#### Shifting MarketCap up for each company (TICKER)

The purpose is to generate the relations of the independent variables with the next year's MARKETCAP. We will use this to then fit the model on just 2013 data for the independent variables, and predict 2014 MARKETCAP for the respective companies, which we can test for precision using scraped Yahoo Finance 2014 current Market Cap. Current Market Cap (December 2014) is highly relevant, as it resembles the end of 10-K cycle MARKETCAP reporting.

```python
df_list = []
for x in ticker_list(sp500):
    df1 = sp500[sp500.TICKER == x]
    df1.MARKETCAP = df1.MARKETCAP.shift(-1)
    df1 = df1[:-1]
    df_list.append(df1)
sp500v2 = pd.concat(df_list) 

df_list = []
for x in ticker_list(N_sp500):
    df1 = N_sp500[N_sp500.TICKER == x]
    df1.MARKETCAP = df1.MARKETCAP.shift(-1)
    df1 = df1[:-1]
    df_list.append(df1)
N_sp500v2 = pd.concat(df_list) 
```

```python
# Eliminating marketcap related variables
# Dividing by ASSETS all variables (including MARKETCAP) except growth variables and ratios
row = len(indicators) - 1
while row >= 0:
    if indicators.Category[row] == "MC":
        sp500v3 = sp500v3.drop(indicators.Indicator[row], axis = 1)
        sp500v2013 = sp500v2013.drop(indicators.Indicator[row], axis = 1)
        row -= 1
    elif indicators.Category[row] == "AS" or indicators.Category[row] == "NORMAL" or indicators.Category[row] == "MARKETCAP": 
        sp500v3[indicators.Indicator[row]] /= sp500v3["ASSETS"]
        sp500v2013[indicators.Indicator[row]] /= sp500v2013["ASSETS"]
        row -= 1
    else: 
        row -= 1

row = len(indicators) - 1
while row >= 0:
    if indicators.Category[row] == "MC":
        N_sp500v3 = N_sp500v3.drop(indicators.Indicator[row], axis = 1)
        N_sp500v2013 = N_sp500v2013.drop(indicators.Indicator[row], axis = 1)
        row -= 1
    else: 
        row -= 1
```

### Finding combinations of Variables Yielding Least Mean Squared Error, sorted

#### Finding combinations for NON-NORMALIZED DATA

```python
# Purpose is to sort the mean squared error combinations of variables, 
# starting with using the top feature, the top and second to top, and so on, for the 20 top features
from sklearn import linear_model
clf = linear_model.LinearRegression()
list_mse = []
for x in range(1,20):
    XX_train = sp500_train[sorted_list_ticker[-x:]].values
    XX_test = sp500_test[sorted_list_ticker[-x:]].values
    YY_train=sp500_train[["MARKETCAP"]].values
    YY_test=sp500_test[["MARKETCAP"]].values
    clf.fit(XX_train, YY_train)
    clf.coef_
    mse = np.mean((YY_test-clf.predict(XX_test)) ** 2)
    list_mse.append([mse,x])
    print ("Test %i: Mean squared error is %.2f" %(x, mse))
```

## SKLEARN REGRESSIONS FOR QUANDL DATA (NON-NORMALIZED AND NORMALIZED)

#### Extracting Current Market Cap Data, to-date 2014, for companies in both QUANDL sets

Current Market Cap Data is relevant since we are in December 2014, which matches the Market Cap which would appear in the majority of 2014 10-K Filings. We are using 2013 annual reports to predict 2014 Current Marketcap / 2013 Assets 

```python
marketcap_list = []
for ticker in sp500v2013.TICKER:
    url_mc = "http://finance.yahoo.com/d/quotes.csv?s=%s&f=j1" %ticker
    temp = list(pd.read_csv(url_mc))
    marketcap_list.append(temp)
marketcap_list[:10]
```

```python
marketcap_column = []
for x in range(0, len(marketcap_list)):
    y = int(float(marketcap_list[x][0][:-1])*(10**9))
    marketcap_column.append(y)
```

```python
#non-normalized
sp500v4 = sp500v2013.copy()
sp500v4 = sp500v4.reset_index()
sp500v4 = sp500v4.drop(sp500v4.columns[0],axis=1)
sp500v4["ASSETS_2013"] = sp500v4['ASSETS']
sp500v4["ActualMC_2014_QUANDL"] = marketcap_column
sp500v4["ActualMC_2013_QUANDL"] = sp500v4["MARKETCAP"]

#normalized
N_sp500v4 = N_sp500v2013.copy()
N_sp500v4 = N_sp500v4.reset_index()
N_sp500v4 = N_sp500v4.drop(N_sp500v4.columns[:2],axis=1)
N_sp500v4["ASSETS_2013"] = N_sp500v4["ASSETS"]
N_sp500v4["ActualMC_2014_QUANDL"] = marketcap_column
N_sp500v4["ActualMC_2013_QUANDL_NORMALIZED"] = N_sp500v4["MARKETCAP"]
```

```python
#running 15 top variable combination regressions and averaging results
#non-normalized
predictions_list = []
for x in range(0,15):
    variables = list_mse[x][1]
    XX_train = sp500_train[sorted_list_ticker[-variables:]].values
    XX_test = sp500_test[sorted_list_ticker[-variables:]].values
    YY_train=sp500_train[["MARKETCAP"]].values
    YY_test=sp500_test[["MARKETCAP"]].values
    clf.fit(XX_train, YY_train)
    rates13 = sp500v4[sorted_list_ticker[-variables:]].values
    sp500v4["Predicted_MC/ASSETS"] = clf.predict(rates13)
    sp500v4["Predicted_MC"] = sp500v4["ASSETS_2013"] * sp500v4["Predicted_MC/ASSETS"]
    prediction = list(sp500v4["Predicted_MC"])
    predictions_list.append(prediction)

#normalized
N_predictions_list = []
for x in range(0,15):
    variables = N_list_mse[x][1]
    XX_train = N_sp500_train[N_sorted_list_ticker[-variables:]].values
    XX_test = N_sp500_test[N_sorted_list_ticker[-variables:]].values
    YY_train=N_sp500_train[["MARKETCAP"]].values
    YY_test=N_sp500_test[["MARKETCAP"]].values
    clf.fit(XX_train, YY_train)
    rates13 = N_sp500v4[N_sorted_list_ticker[-variables:]].values
    N_sp500v4["Predicted_MC"] = clf.predict(rates13)
    prediction = list(N_sp500v4["Predicted_MC"])
    N_predictions_list.append(prediction)
```

```python
#non-normalized
df_avg_predictions = []
for x in range(0,len(sp500v4)):
    temp = 0
    for y in range (0,15):
        temp += predictions_list[y][x]
    avg = temp / 15
    df_avg_predictions.append(avg)
print df_avg_predictions[:5]

#normalized
N_df_avg_predictions = []
for x in range(0,len(N_sp500v4)):
    temp = 0
    for y in range (0,15):
        temp += N_predictions_list[y][x]
    avg = temp / 15
    N_df_avg_predictions.append(avg)
print N_df_avg_predictions[:5]
```

```python
#Creating Results for Non-Normalized
sp500v4["PredictedMC_2014"] = df_avg_predictions
sp500v4["Actual_PctGrowthMC"] = (sp500v4["ActualMC_2014_QUANDL"] - (sp500v4["ActualMC_2013_QUANDL"]*sp500v4["ASSETS_2013"]))/ (sp500v4["ActualMC_2013_QUANDL"]*sp500v4["ASSETS_2013"])*100
sp500v4["Predicted_PctGrowthMC"] = (sp500v4["PredictedMC_2014"] - (sp500v4["ActualMC_2013_QUANDL"]*sp500v4["ASSETS_2013"]))/ (sp500v4["ActualMC_2013_QUANDL"]*sp500v4["ASSETS_2013"])*100
sp500v5 = sp500v4[['TICKER','ASSETS_2013','ActualMC_2013_QUANDL','PredictedMC_2014','ActualMC_2014_QUANDL','Predicted_PctGrowthMC','Actual_PctGrowthMC']]
sp500v5.head()
```

```python
#Creating Results for Normalized
N_sp500v4["PredictedMC_2014"] = N_df_avg_predictions
N_sp500v4["Normalized_ActualMC_2014"] = N_sp500v4["ActualMC_2014_QUANDL"] / N_sp500v4["ASSETS_2013"]
mc_mean_2014 = N_sp500v4["Normalized_ActualMC_2014"].mean()
N_sp500v4["Normalized_ActualMC_2014"] -= mc_mean_2014
N_sp500v4["Actual_PctGrowthMC_NORMALIZED"] = ((N_sp500v4["Normalized_ActualMC_2014"] - N_sp500v4["ActualMC_2013_QUANDL_NORMALIZED"]) / abs(N_sp500v4["ActualMC_2013_QUANDL_NORMALIZED"]))*100
N_sp500v4["Predicted_PctGrowthMC_NORMALIZED"] = ((N_sp500v4["PredictedMC_2014"] - N_sp500v4["ActualMC_2013_QUANDL_NORMALIZED"]) / abs(N_sp500v4["ActualMC_2013_QUANDL_NORMALIZED"]))*100
N_sp500v5 = N_sp500v4[['TICKER','ActualMC_2013_QUANDL_NORMALIZED','PredictedMC_2014','Normalized_ActualMC_2014','Predicted_PctGrowthMC_NORMALIZED','Actual_PctGrowthMC_NORMALIZED']]
N_sp500v5.head()
```

```python
#Precision in non-normalized data
Precision_NonNormalized = 0
for x in range(0,len(sp500v5)):
    actual = sp500v5['ActualMC_2014_QUANDL'][x]
    if actual == 0:
        Precision_NonNormalized += 0
    else:
        temp = (sp500v5['PredictedMC_2014'][x] - actual) / actual
        temp = temp**2
        Precision_NonNormalized += temp 

print "The Precision Metric of the Non-Normalized Regression is : %s"%Precision_NonNormalized

#Precision in normalized data
Precision_Normalized = 0
for x in range(0,len(N_sp500v5)):
    actual = N_sp500v5['Normalized_ActualMC_2014'][x]
    if actual == 0:
        Precision_Normalized += 0 
    else:
        temp = (N_sp500v5['PredictedMC_2014'][x] - actual) / actual
        temp = temp**2
        Precision_Normalized += temp 

print "The Precision Metric of the Normalized Regression is : %s"%Precision_Normalized
```

The Lines below impute missing data using a mean of column strategy. After doing this, the normalize function does two things 1) divide all the data by the value of "Assets" and 2) normalizes the data by subtracting the value from the mean of the column. The order for this is 1. Impute 2. Divide By Assets 3. Impute. We will see below that dividing by Assets renders SVD unusable in terms of dividing the data set into multiple classes. We will use svd at the end of the notebook, but the cost of this is that we will not divide by assets.

```python
def imputeData(data):
    for col,val in data.iteritems():
        if ("year" not in col.lower()) & ("ticker" not in col.lower()):
            data[col].fillna(data[col].mean(), inplace=True)

# Divide By Assets And Normalize Missing Data With Mean
def normalize(data,normalize_only):
    for col,val in data.iteritems():
        # Divide By Assets
        if not normalize_only:
            if ("year" not in col.lower()) & ("ticker" not in col.lower()):
                if "assets" == col.lower():
                    continue
                for i in range(0,len(data[col])):
                    asset = 1
                    try:
                        asset = data["Assets"].iloc[i]
                    except:
                        asset = data["ASSETS"].iloc[i]
                    data[col].iloc[i] = data[col].iloc[i] / asset
        # Normalize
        elif normalize_only:
            if "assets" == col.lower():
                continue
            if ("year" not in col.lower()) & ("ticker" not in col.lower()):
                for i in range(0,len(data[col])):
                    data[col].iloc[i] = data[col].iloc[i] - (data[col].mean())
```

### Optimizing the 'k' in KMeans
In order to select the optimum number of classes for both of our datasets, we need to investigate how changing the value of k will impact our clusters. We do this by employing the 'elbow' approach, which is to say, look at how close all of our data points are (sum of squares distance) to the centroid predicted for each point by a model with one particular k value.

```python
from scipy.cluster.vq import kmeans

# Vector Quantization
max_k = 25
k_list = list(range(2,max_k))

# http://stackoverflow.com/questions/6645895/calculating-the-percentage-of-variance-measure-for-k-means
# Determine Optimal K By Calculating Sum Of Squares (Inspired by SarahGuido's tutorial)
def plotSOS(dim):
    soss = []
    for k in range(2,max_k):      
        centroid = kmeans(dim,k)[0]
        distance = dst.cdist(dim, centroid, 'euclidean')
        min_dist = np.mean(distance,axis=1)
        sos = sum(min_dist)/dim.shape[0]   
        soss.append(sos)
    return soss
        
def showSOSPlot(ddata, chart_title):
    plt.figure(figsize=(8, 6))
    plt.plot(k_list, ddata, linestyle='--', marker='o', color='blue')
    plt.xlabel('K Value')
    plt.ylabel('Mean Sum Of Squares Euclidian Distance To Centroid')
    plt.title(chart_title)
    plt.show()

s_soss = plotSOS(scrapedTrain)
f_soss = plotSOS(ratiosTrain)
showSOSPlot(s_soss, "Scraped Data K Values vs. Mean Sum Of Squares Distance")
showSOSPlot(f_soss, "Ratios Data K Values vs. Mean Sum Of Squares Distance")
```

We find that our ratios dataset very neatly fits the shape of the 'elbow', and with incremental increase in the value of k, the distance between points and centroid decreases, which is to say, the higher the k value, the better fitted the data. However, as theorized in the 'elbow' approach, each k value demonstrates a diminishing return in terms of improvement in proximity between data and centroid. The scraped dataset forms a very rough elbow, but only after class 7, with k= ~7 being a good number in terms of marginal improvement in cluster density. For now, we will say, that k value 4-10 seems like a good starting point.
<p> The scraped data clearly suffers when k values are greater than 3-4, only recovering when the number of classes approaches 25.</p>

#### Using DPGMM Classifier and 'elbow' k value to guess number of classes

The DPGMM classifier takes a k value as a maximum number of componenents, and uses this maximum value to guide the number of classes it ends up selecting through the dirichlet process. We have used the elbow value, a number roughly 4-10, to guide the number of classes selected. We run the classifier maximum number of classes 4-10. We vary the alpha, which determines how many components are used to fit the data. The higher the alpha, the more likely the distribution of points is to be even among the classes.

```python
# Test DPGMM for different alpha parameters, and set maximum number of componenets at 15.
def runUntrained(b,trainDPG, testDPG,dpg_title,model):

    model.fit(trainDPG)
    res = model.predict(testDPG)
    plt.figure(figsize=(8, 4))
    plt.hist(res, bins=b+1, width=0.3)
    plt.xlabel('Class')
    plt.ylabel('Class Frequency')
    plt.title(dpg_title)
    plt.show()
    return res

a = 0.01
while a < 100.0:
    dpgModel = sklearn.mixture.DPGMM(n_components=15, covariance_type='diag', alpha=a, 
                                 random_state=None, thresh=0.01, verbose=False, min_covar=None, 
                                 n_iter=10, params='wmc', init_params='wmc')
    res_scr = runUntrained(15,scrapedTrain, scrapedTest, "alpha = " + str(a) + ' Scraped Data Class Frequencies DPGMM Model',dpgModel)
    res_ratios = runUntrained(15,ratiosTrain, ratiosTest, "alpha = " + str(a) + ' Ratios Data Class Frequencies DPGMM Model',dpgModel)
    a *= 10
```

The scraped dataset is too monolithic: we are inclined to forget this dataset entirely, regardless of the amount of work put into assembling it.

The ratios dataset, however, clusters into three to four classes, with one of these dominant in a major way, a second and a third much less significant, and the rest containing a tiny fraction of the data. Since DPGMM tends not to respond to much to the parameters (we saw this when we ran multiple models with different alpha values), we can reasonably assume that the ratios dataset is clustered along the following lines: one diminant class, and two minority classes.

Perhaps this is because we are using too much data, or that we may be using the wrong algorithm for unsupervised learning. We will go back to our KMeans, and re-run the model to optimize the number of clusters.

```python
from hcluster import pdist, linkage, dendrogram

# Generate Dendrogram to Show Class Relationships
def generateDendro(X,Y,dtitle):
    classes = {}

    for i in range(0,len(Y)):
        if str(Y[i]) in classes:
            classes[str(Y[i])] += X[i]
        else:
            classes[str(Y[i])] = X[i]

    class_keys = []
    class_vals = []
    for key,val in classes.iteritems():
        class_keys.append(key)
        class_vals.append(val)


    plt.figure(figsize=(8, 6))
    plt.xlabel('Class #')
    plt.ylabel('Class Distance')
    plt.title(dtitle)
    class_arr = np.array(class_vals)
    class_distances = pdist(class_arr)
    class_link = linkage(class_distances)
    dendrogram(class_link, labels=class_keys)
    plt.show()
```

```python
def runUntrained(b,trainModel, testModel,chart_title,model):
    model.fit(trainModel)
    res = model.predict(testModel)
    plt.figure(figsize=(8, 4))
    plt.hist(res, bins=b, width=0.3)
    plt.xlabel('Class #')
    plt.ylabel('Class Frequency')
    plt.title(chart_title)
    plt.show()
    return res


results_ratios = []
results_scraped = []

for k in range(3,8):
    kmeansModel = sklearn.cluster.KMeans(n_clusters=k, init='k-means++', n_init=10, tol=0.01, precompute_distances=True)
    res_scr = runUntrained(15,scrapedTrain, scrapedTest, "K = " + str(k) + ' Scraped Data Class Frequencies KMeans Model',kmeansModel)
    generateDendro(scrapedTest,res_scr,"Scraped data distribution of classes by KMeans when k = " + str(k))
    res_ratios = runUntrained(15,ratiosTrain, ratiosTest, "K = " + str(k) + ' Ratios Data Class Frequencies KMeans Model',kmeansModel)
    generateDendro(ratiosTest,res_ratios,"Ratios data distribution of classes by KMeans when k = " + str(k))
    results_ratios.append(res_ratios)
    results_scraped.append(res_scr)
```

### Test Number of Classes Without Specifying Upper Bound Using MeanShift
We will now test the results we got in terms of the number of classes using MeanShift, which, like DPGMM, will automatically select the number of classes, although unlike DPGMM, there will be no need to specify an upper bound.

```python
from sklearn.cluster import MeanShift, estimate_bandwidth


def predictMeanShift(test,train,dendro_title,hist_title,nsamples):
    # We will leave bandwidth as blank so that mean_shift can estimate it one for us
    bandwidth = estimate_bandwidth(train, quantile=0.2, n_samples=nsamples)

    # Not supplying bandwidth leads to one class, which means that he automatically assigned bandwidth is too large
    meanShift = MeanShift(bandwidth=bandwidth, bin_seeding=True)

    meanShift.fit(train)

    labels = meanShift.fit_predict(test)
    labels_unique = np.unique(labels)

    ms_res = meanShift.cluster_centers_
    generateDendro(test,labels,dendro_title)

    plt.figure(figsize=(8, 4))
    plt.hist(labels, bins=len(labels_unique), width=0.3)
    plt.xlabel('Class #')
    plt.ylabel('Class Frequency')
    plt.title(hist_title)
    plt.show()
    return labels

num_samples = 20
while num_samples < 150:
    ratios_meanshift_res = predictMeanShift(ratiosTest,ratiosTrain,
                                         "Ratios Data MeanShift Class Distribution When Samples = "+ str(num_samples),
                                         "Ratios Data Class frequencies MeanShift When Samples = " + str(num_samples),num_samples)
    num_samples += 50
```

KMeans reveals a number of classes that correlates with the sample size. As the sample size increases, so do the number of classes. These classes are relatively evenly distributed among 4-5 major classes, and a large number of minor ones. The dendrogram reveals that one class (class number varies in every model), which seems to be simplicifolious, very likely contains outliers, and that the rest of the classes are all very similar to each other, hence the tiny vertical distance between the vast majority of the leaves. The meanshift dendrogram, like the ones above, show two principal clades, one of them containing many closely related clades, and the other one with a lone leaf. It seems reasonable to conclude that most companies in the S&P 500 flock together in terms of financials.

### How Many Components Should We Use?
We need to figure out how many components (columns) are optimal. For this, SVD will have come in handy. We run kmeans with a k value of 7 because this was approximately our 'elbow' value below, varying the components we take from the data from 2 to 20.

```python
def getSumOfSquare(dim):    
    centroid = kmeans(dim,4)[0]
    distance = dst.cdist(dim, centroid, 'euclidean')
    min_dist = np.mean(distance,axis=1)
    sos = np.mean(sum(min_dist)/dim.shape[0])
    return sos

sumOfSquares_scr = []
sumOfSquares_ratios = []
numClust = []

for n in range(1,20):
    numClust.append(n)
    sosList_scr = []
    sosList_ratios = []
    for i in range(0,1):

        scr_tr = getSumOfSquare(raTrain[:n])
        ratios_tr = getSumOfSquare(raTest[:n])

        sosList_scr.append(scr_tr)
        sosList_ratios.append(ratios_tr)
    
    sumOfSquares_scr.append(np.mean(sosList_scr))
    sumOfSquares_ratios.append(np.mean(sosList_ratios))
    
plt.figure(figsize=(15, 6))
plt.plot(numClust, sumOfSquares_scr, linestyle='--', marker='o', color='blue')
plt.xlabel('Number Of Components')
plt.ylabel('Mean Sum Of Squares Euclidian Distance To Centroid')
plt.title("Scraped Data: Number Of Components Vs. Mean Sum Of Squares Distance To Centroid")
plt.show()

plt.figure(figsize=(15, 6))
plt.plot(numClust, sumOfSquares_ratios, linestyle='--', marker='o', color='red')
plt.xlabel('Number Of Components')
plt.ylabel('Mean Sum Of Squares Euclidian Distance To Centroid')
plt.title("Ratios Data: Number Of Components Vs. Mean Sum Of Squares Distance To Centroid")
plt.show()
```

Clearly, minimizing the number of components in the ratios dataset will increase the density of our clusters. We will therefore use the first five columns of our data, since after this, the density decreases radically.

### Testing Model Consistentcy Using Rand, and Other Metrics
To ensure that we select a good model, we need to test how consistently it will produce the same classes. We will test multiple models using the rand, homogeneity, and completeness scores.

```python
import sklearn.metrics as metrics

def getScores(true,predicted,modelName):
    print modelName
    print "Addjusted rand score: " + str(metrics.adjusted_rand_score(true,predicted))
    print "Homogeneity score: " + str(metrics.homogeneity_score(true, predicted))
    print "Completeness Score: " + str(format(metrics.completeness_score(true, predicted)))
    print "___________________________________________________"

def execMeanShift(train,test):
    bandwidth = estimate_bandwidth(train, quantile=0.2, n_samples=50)
    meanShift = MeanShift(bandwidth=bandwidth, bin_seeding=True)
    meanShift.fit(train)
    res = meanShift.fit_predict(test)
    return res

def execDPGMM(i,train,test):
    dpgModel = sklearn.mixture.DPGMM(n_components=i, covariance_type='diag', alpha=1.0, 
                                 random_state=None, thresh=0.01, verbose=False, min_covar=None, 
                                 n_iter=10, params='wmc', init_params='wmc')
    dpgModel.fit(train)
    res = dpgModel.predict(test)
    return res

def execKMeans(k,train,test):
    kmModel = sklearn.cluster.KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=300, tol=0.0001, precompute_distances=True, verbose=0, random_state=None, copy_x=True, n_jobs=1)
    kmModel.fit(train)
    res = kmModel.predict(test)
    return res
        
#Test MeanShift Consistency
for x in range(0,5):
    labels = execMeanShift(ratiosTrain,ratiosTest)
    labels2 = execMeanShift(ratiosTrain,ratiosTest)
    getScores(labels,labels2,"MeanShift Result")  
    
# Test DPGMM Consistency
for i in range(5,11):
    dpg_res = execDPGMM(i,ratiosTrain,ratiosTest)
    dpg_res2 = execDPGMM(i,ratiosTrain,ratiosTest)
    getScores(dpg_res,dpg_res2,"DPGMM Result When Components = " + str(i))    

# Test KMeans Consistency
for k in range(2,8):
    result = execKMeans(k,ratiosTrain,ratiosTest)
    result2 = execKMeans(k,ratiosTrain,ratiosTest)
    getScores(result2,result,"KMeans Result When K = " + str(k))    
```

```python
#Define functions here
"""
Queries the Quandl API with the given url and returns the result
Parameters: Quandl API url
Return: (True, Quandl resultset) or (False, None)
"""
def getQuandlData(url):
    #Query the fundamental data using Quandl API
    try:
        qldata = ql.get(url, authtoken=qltoken)
        return 1, qldata
    except:
        return 0, None
    
"""
Function to download the file from a given URL
Return: File handle
"""
def downloadFile(url):
    source =  StringIO.StringIO(requests.get(url).content)
    return source
    
#using the function to plot the decision surface from HW5 for helping in visualizing the classifier performance
"""
Plot the decision surface of a classifier
"""
def plot_decision_surface(clf, X_train, Y_train, X_label, Y_label):
    plot_step=0.1
    
    if X_train.shape[1] != 2:
        raise ValueError("X_train should have exactly 2 columnns!")
    
    x_min, x_max = X_train[:, 0].min() - plot_step, X_train[:, 0].max() + plot_step
    y_min, y_max = X_train[:, 1].min() - plot_step, X_train[:, 1].max() + plot_step
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))

    clf.fit(X_train,Y_train)
    if hasattr(clf, 'predict_proba'):
        Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:,1]
    else:
        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])    
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Reds)
    plt.scatter(X_train[:,0],X_train[:,1],c=Y_train,cmap=plt.cm.Paired)
    plt.xlabel(X_label)
    plt.ylabel(Y_label)
    plt.title(clf)
    plt.show()    
```

```python
#Use the fundamental data collected from Quandl, that is normalized
fundamentals = pd.read_csv(downloadFile(fundamentals_url), index_col=None)
print "Total records available in Quandl dataset:", len(fundamentals)
fundamentals.head()
```

**Define X and Y for Classification**

Before analysis, seperate out the fundamentals variables that are unrelated to Market Capitalization as input X. The Market Capitalization is used as the Y and it is classified into a binary variable, above market average as '1' and below market average as '0'.

```python
#Define the Marketcap as above average or below average
def map_marketcap(marketcap):
    if marketcap >= 0:
        return 1
    elif marketcap < 0:
        return 0
    else:
        return None

#Define the X and Y for classification
fundamentalssup = fundamentals[['ACCOCI', 'ASSETSC', 'ASSETSNC', 'CASHNEQ', 'DEBT', 'EQUITY', 'LIABILITIES', 'LIABILITIESC', 'LIABILITIESNC', 'CAPEX', 'DEPAMOR', 'NCF', 'NCFF', 'NCFI', 'NCFO', 'NCFX', 'DPS', 'EBIT', 'INTEXP', 'NETINC', 'NETINCCMN', 'NETINCDIS', 'PREFDIVIS', 'REVENUE', 'TAXEXP']]
fundamentalssup['MARKETCAPBINARY'] = fundamentals['MARKETCAP'].map(map_marketcap)
fundamentalssup.head()

#Store the X and Y to a numpy array
Ysup = np.array(fundamentalssup['MARKETCAPBINARY'].values)
fundamentalssup = fundamentalssup.drop(['MARKETCAPBINARY'], axis=1)
Xsup = fundamentalssup.as_matrix()

#print the length and the X and Y
print len(Xsup), len(Ysup)
print Xsup, Ysup

#Percentage of data with above average Marketcap
print "\nPercentage of dataset with above average Marketcap ", round(np.bincount(Ysup)[1] * 100 / len(Ysup), 2), "%"
```

**Principal Component Analysis**

Perform SVD for dimenesionality reduction on the X variables and do PCA on the top two components.

```python
#Perform PCA on the fundamentals dataset for dimensionality reduction
pca = PCA(n_components=2)
Xpca = pca.fit_transform(Xsup)

#Plot the first two Principal Components from PCA analysis
plt.figure(figsize=(12,8))
plt.scatter(Xpca[:,0], Xpca[:,1], c=Ysup, s = 10, cmap=plt.cm.prism)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('First two Principal Components from the fundamentals dataset')
plt.show()

#Print the PCs
print len(Xpca)
print Xpca
```

**Result**

PCA did not help in the class separability of the fundamental variables as can be seen, a large chunk of the results are still overlapping.

**Classification Method 1 - Random Forests**

Apply Random Forest techniques and train the classifier. Initially, an optimal number of decision tree is determined. Then the important features of classification are identified. 

```python
#Split the train and test data
Xsup_train, Xsup_test, Ysup_train, Ysup_test = sklearn.cross_validation.train_test_split(Xsup, Ysup, test_size=0.3, random_state=19)
print Xsup_train.shape, Ysup_train.shape, Xsup_test.shape, Ysup_test.shape

#Run the Random Forest Classifier for varying number of trees to find the optimal decision trees
all_scores = []
for i in range(1,20):
    forest = sklearn.ensemble.RandomForestClassifier(n_estimators=i)
    scores = sklearn.cross_validation.cross_val_score(forest, Xsup_train, Ysup_train, cv=10)   
    all_scores.append(scores)
    
#Plot the results
plt.figure(figsize=(18,10))
sns.boxplot(all_scores,names = [i for i in range(1,20)])
#sns.violinplot(all_scores, names = [i for i in range(1,20)], color="Paired", bw=0.5);
plt.ylim(ymin=0.5, ymax=1.0)
plt.ylabel("Score")
plt.xlabel("Number of Decision Trees")
plt.title('Scores of Random Forests vs Number of Decision Trees')
plt.show()    
```

```python
#Train the classifier again with the optimal no. of decision trees
forest = sklearn.ensemble.RandomForestClassifier(n_estimators=8)
forest.fit(Xsup_train, Ysup_train)
importance = forest.feature_importances_

#Store the results of the classifier
forestresults = []
forestresults = pd.DataFrame(columns= ['features','score'])
for i in range(len(importance)):
    forestresults.at[i, 'features'] = fundamentalssup.columns[i]
    forestresults.at[i, 'score'] = importance[i]
    
#Feature Selection
forestresults = forestresults.sort(forestresults.columns[1], ascending=False)    
forestresults = forestresults.reset_index()
print forestresults
```

```python
#Plot the results of the feature selection from Random Forest
plt.figure(figsize=(12,8))
sns.barplot(forestresults['features'], y=forestresults['score'], x_order=forestresults['features'], palette="BuGn_d")
plt.ylim(ymin=0.0, ymax=0.25)
plt.xticks(rotation=90)
plt.ylabel("Feature Importance")
plt.xlabel("Feature")
plt.title('Features importance in Marketcap classification (Random Forest)')
plt.show()       
```

**Classification Method 2 - Support Vector Machines**

Apply Support Vector Machine techniques and train the classifier. 

```python
#Use Support Vector Machine as a classifier
svm = sklearn.svm.SVC(C=1, gamma=0.5, probability=True, class_weight='auto', kernel='linear')
svm.fit(Xsup_train, Ysup_train)
coef = svm.coef_

#Store the results of the SVM classifier
svmresults = []
svmresults = pd.DataFrame(columns= ['features','weights'])
for i in range(len(coef[0])):
    svmresults.at[i, 'features'] = fundamentalssup.columns[i]
    svmresults.at[i, 'weights'] = coef[0][i]
    
#Feature Selection
svmresults = svmresults.sort(svmresults.columns[1], ascending=False)    
svmresults = svmresults.reset_index()
print svmresults
```

```python
#Plot the results of the feature selection from SVM
plt.figure(figsize=(12,8))
sns.barplot(svmresults['features'], y=svmresults['weights'], x_order=svmresults['features'], palette="BuGn_d")
plt.xticks(rotation=90)
plt.ylabel("Feature Weight")
plt.xlabel("Feature")
plt.title('Feature weights in Marketcap classification (SVM)')
plt.show()   
```

# QUANDL DATA DIVIDED BY ASSETS (NOT NORMALIZED)

### Loading Data: Divided by ASSETS

```python
fundamentals_url = '0BzgUPv-n9Oh3MVdsWF9aZTB5eFk'
sp500_orig = pd.read_csv(downloadFile(drive_url+fundamentals_url))
# keeping sp500_orig for later use
sp500 = sp500_orig.copy()
```

```python
file_categories = "0BzgUPv-n9Oh3c2ppdFJURDFLeXM"
indicators = pd.read_csv(downloadFile(drive_url+file_categories))
```

```python
# sp500 (first version) eliminates 2014 data points 
sp500 = sp500[sp500.YEAR != 2014].reset_index()
sp500 = sp500.drop(sp500.columns[0], axis=1)
```

```python
# Function that gives you a list of the dataframe's current tickers
def ticker_list(df):
    x = 1
    tickers=[df.TICKER[0]]
    while x < len(df):
        if df.TICKER[x]!= df.TICKER[x-1]:
            tickers.append(df.TICKER[x])
            x+=1
        else:
            x+=1
    return tickers
```

```python
ticker_list(sp500)[:10]
```

```python
df_list = []
for x in ticker_list(sp500):
    df1 = sp500[sp500.TICKER == x]
    df1.MARKETCAP = df1.MARKETCAP.shift(-1)
    df1 = df1[:-1]
    df_list.append(df1)
sp500v2 = pd.concat(df_list) 
```

```python
# sp500v2 includes a shift in MARKETCAP up for each company, eliminating latest year's data
sp500v2.tail()
```

```python
sp500v3 = sp500v2.copy()
#sp500v2013 will be used later to make predictions for next year's MarketCap based on current annual data
sp500v2013 = sp500[sp500.YEAR == 2013]
#to be used for adjusting Actual MC 2014 to Actual MC 2014 / ASSETS 2013 later 
sp500v2013_ASSETS = sp500v2013['ASSETS']

# Eliminating marketcap related variables
# Dividing by ASSETS all variables (including MARKETCAP) except growth variables and ratios
row = len(indicators) - 1
while row >= 0:
    if indicators.Category[row] == "MC":
        sp500v3 = sp500v3.drop(indicators.Indicator[row], axis = 1)
        sp500v2013 = sp500v2013.drop(indicators.Indicator[row], axis = 1)
        row -= 1
    elif indicators.Category[row] == "AS" or indicators.Category[row] == "NORMAL" or indicators.Category[row] == "MARKETCAP": 
        sp500v3[indicators.Indicator[row]] /= sp500v3["ASSETS"]
        sp500v2013[indicators.Indicator[row]] /= sp500v2013["ASSETS"]
        row -= 1
    else: 
        row -= 1
        
sp500v2013 = sp500v2013.drop('ASSETS', axis = 1)
```

```python
# Adding the gdp growth variable to a new column
file_GDPGROWTH = "0BzgUPv-n9Oh3YldUVXo1Y2pabkU"
gdpgrowth = pd.read_csv(downloadFile(drive_url+file_GDPGROWTH))

sp500v3 = sp500v3.merge(gdpgrowth, how='outer', on="YEAR")
sp500v3 = sp500v3.sort(['TICKER','YEAR']).reset_index()
sp500v3 = sp500v3.drop(sp500v3.columns[0], axis=1)

#Add 2013 GDP growth
sp500v2013['GDP_GROWTH'] = 1.9

# to account for addition of another row
sp500v3 = sp500v3[:-1]
```

```python
year_list = []
list_tickers = []
for x in edgar.CoYear:
    year = int(x[-4:])
    year_list.append(year)
    ticker = x[:-5]
    list_tickers.append(ticker)
```

```python
edgar["YEAR"] = year_list
edgar["TICKER"] = list_tickers
edgar.head()
```

```python
edgar = edgar[edgar.YEAR != 2014].reset_index()
edgar = edgar.drop(edgar.columns[0], axis=1)
```

```python
x = 0
lst_newrows = []
lst_delete = []
while x < len(edgar):
    if edgar.CoYear[x] == "A_2009":
        x+=1
    elif (edgar.YEAR[x] != (edgar.YEAR[x-1] + 1)) & (edgar.TICKER[x] == edgar.TICKER[x-1]):
        if edgar.YEAR[x] >= edgar.YEAR[x-1] + 3:
            lst_delete.append(edgar.CoYear[x-1])
        else:
            temp_lst = []
            for column in edgar.columns:
                if type(edgar[column][x]) == np.float64:
                    value = (edgar[column][x] + edgar[column][x-1])/2
                    temp_lst.append(value)
                elif column == 'CoYear':
                    temp_lst.append('%s_%i'%(edgar['TICKER'][x],int(edgar['YEAR'][x]-1)))
                elif column == 'YEAR':
                    temp_lst.append(edgar[column][x] - 1)
                else:
                    value=edgar[column][x]
                    temp_lst.append(value)
            dfappend = pd.DataFrame(temp_lst).T
            lst_newrows.append(dfappend)
        x+=1
    else:
        x+=1
```
